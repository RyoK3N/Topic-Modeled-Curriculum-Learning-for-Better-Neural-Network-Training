```python
import os
from datasets import load_dataset

base_dir = "data/experiment/101"
os.makedirs(base_dir, exist_ok=True)

datasets_to_download = {
    "cifar10": "cifar10",
    "cifar100": "cifar100",
    "mnist": "mnist",
    "fashion_mnist": "fashion_mnist",
    "ag_news": "ag_news",
    "imdb": "imdb"
}

for name, hf_id in datasets_to_download.items():
    cache_dir = os.path.join(base_dir, name)
    os.makedirs(cache_dir, exist_ok=True)
    
    print(f"Downloading {name}...")
    _ = load_dataset(hf_id, cache_dir=cache_dir)
    print(f"✓ {name} saved to {os.path.abspath(cache_dir)}")

print(f"\n✅ All datasets downloaded under: {os.path.abspath(base_dir)}")
```
